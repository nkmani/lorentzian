```python
class Settings:
    # Default settings defined as class attributes
    font_size = 12
    theme = "dark"
    debug_mode = False
    plugins = [] # Caution: Mutable defaults can be tricky if modified in place

    def __init__(self, **kwargs):
        # Set instance attributes from kwargs, overriding defaults
        for key, value in kwargs.items():
            setattr(self, key, value)

    def get_changed_settings(self):
        # Get instance attributes
        instance_attributes = vars(self)
        # Get class attributes
        class_attributes = vars(type(self))

        changed_settings = {}
        for key, value in instance_attributes.items():
            # Check if the attribute is a setting we want to track (e.g., starts with a letter)
            # and if it differs from the class default
            if not key.startswith('__') and key in class_attributes:
                if value != class_attributes[key]:
                    changed_settings[key] = value

        return changed_settings

# Create an instance with some default settings overridden
user_settings = Settings(font_size=16, theme="light", debug_mode=False)

# List the changes
changes = user_settings.get_changed_settings()
print("Changed settings:", changes)

# Example with an instance that has no changes from defaults
default_settings = Settings()
default_changes = default_settings.get_changed_settings()
print("Default settings changes:", default_changes)
```
